## Introduction

<p style="text-align: justify;">From voice-powered personal assistants like Siri and Alexa, to more underlying and fundamental technologies such as behavioral algorithms, suggestive searches and self-driving vehicles boasting powerful predictive capabilities, there are quite a few examples and applications of artificial intelligence in use today. Many fear for their jobs or livelihoods, assuming that the rise of AI will displace and replace human labour, rather than complement it. Thankfully, economic theory disputes this. In their recent [article](https://www.nber.org/papers/w24626), Agrawal, Gans and Goldfarb (2018) theorise how prediction by artificial intelligence and judgement by humans  interact in situations where the two can compliment one another. They do so by building a simple model, in which they demonstrate how different scenarios imply substitutiability versus complimentarity for humans and machines. They find that in almost all situations, humans will be able to provide a valuable contribution alongside qualities provided by machines.</p>

<p style="text-align: justify;">To explore the interaction between prediction and judgement, the authors argue that advances in the ability of machines to conduct mental tasks are driven by improvements in machine prediction. Therefore, they examine sources of comparative advantage in the presence of improved machine prediction. Agrawal et al., make a clear distinction between judgement and prediction in their paper, where judgement is classified as an activity that can improve decision-making outcomes. Moreover, prediction is information about the expected state of the world that can be easily described, whereas judgement relies on factors that are indescribable. Next to this, the authors emphasize that judgement is not a passive activity and requires deliberative cognitive application. Adopting the above mentioned distinction, the paper demonstrates several outcomes; it shows that in the absence of prediction, judgement might push decisions towards riskier actions if it identifies hidden opportunities, however, when there is prediction, the reverse is true. Furthermore, Agrawal et al., show that it can often be the case that better judgement will tend to push the design of prediction technology towards choosing more situations where prediction is used, at the expense of those predictions being less reliable. </p>

## Theoretical Background


<p style="text-align: justify;"> Agrewal et al., create a theoretical model in which they explore the consequences of computer prediction and human judgement. To do this adequately using simple equations, they make a number of assumptions, crucially assuming predictable characteristics can only take on 2 forms, R and r, and that the average of these must be lower than the value of the safe action. To help in thinking about this, we envision the casus as deciding on whether to apply an experimental chemo treatment to cancer patients. The outside option is said to be relatively safe, whereas treatment can have a good or bad effect depending on patient characteristics.</p>

$$\frac{1}{2}(R + r) < S$$

<p style="text-align: justify;">From this, it follows that the default action in their case is to choose the safe option (S, e.g. not treating) for each agent presented to them.</p>

<p style="text-align: justify;">They then elaborate on this in their first equation, measuring the average payoff when bringing only computerised prediction into the equation. Here, a successful and positive estimate of the predictable characteristics prompts the algorithm to treat a patient. Conversely, when a prediction is absent, or it identifies negative characteristics, the algorithm chooses to not treat.</p>

$$π_p= e(\frac{1}{2}R + \frac{1}{2}S) + (1 - e)S = e\frac{1}{2}R + ( 1 - e\frac{1}{2})S$$

<p style="text-align: justify;">They then bring in additional patient characteristics, which can be identified solely through human judgement. We refer to these as opportunities and costs. Again, they restrict the values this can take on, and identify it as a single value, which can be positive or negative. Crucially, the average payoff of treating in the presence of a positive judgeable characteristic is greater than choosing the safe option, while a positive state in predictable characteristics combined with a negative state in judgeable characteristics is smaller than the safe payoff.</p>

$$\frac{1}{2}(R + \Delta) + \frac{1}{2}r > S$$

$$R - \Delta < S$$

<p style="text-align: justify;">When only applying human judgement, the safe option again forms the default, where treatment is only applied in the presence of a positive judgeable characteristic. This means negative judgeable characteristics (costs) play a much smaller role than opportunities, and we refer to judgeable characteristics in general as opportunities throughout this paper. This leads to the following average payoff:</p>

$$π_g=\lambda_g \frac{1}{2}\rho (\frac{1}{2}(R + \Delta) + \frac{1}{2}r) + (1 - \lambda_g\frac{1}{2}\rho)S$$

<p style="text-align: justify;">Combining both prediction and judgement allows for a more finetuned approach, where we can combine the outcomes found under π^p and π^j. This way, we benefit from situations in which either method discovers positive attributes. Additionally, we can exclude those situations where we have conflicting information on predictable characteristics and judgeable opportunities, and avoid being overconfident and treating these cases. This allows us to both increase the amount of good states and decrease the amount of bad states we treat, leading to a two-channeled improvement over applying singular methods.</p>

$$π_c = e(\frac{1}{2}(\lambda_b(\frac{1}{2}\rho S+ \frac{1}{2}\rho(R + \Delta)+
 (1-\rho)R) + (1 - \lambda_b)R) + \frac{1}{2}S) + (1-e)(\lambda_g \frac{1}{2}\rho(\frac{1}{2}(R + \Delta) + \frac{1}{2}r) + (1-\lambda_g \frac{1}{2}\rho)S)$$

<p style="text-align: justify;">Agrewal et al go significantly deeper into this, and explore various other scenarios. In this paper, however, we chose to explore a relaxation of some of the core assumptions in the model above. Specifically, we relax the assumption of a six state world [$^1$](#Footnotes) and allow for almost unlimited variations of predictable and judgeable characteristics. We then explore how variations in prediction and judgement success rates influence the average payoffs of the three different decision options ($π_p$, $π_j$ and $π_c$). While we explore this for one set distribution of patient characteristics, we allow the reader to use our model and vary this distribution, the value of the safe treatment and other characteristics present in the model.</p>

```python
#First we choose some variables to run most of the experiment with. 
#You may alter these as you wish, but recommend using the ones provided for the first run.

Seed = 1984
n = 30000
N = 100

mu1 = -0.05
sigma1 = 5

low = -5
high = 5

Safe = 0
```

<p style="text-align: justify;">We first run a test for a single distribution, allowing us to test whether our set up works. </p>

<p style="text-align: justify;">First, we set up a normal distribution for underlying, predictable human characteristics, which can be utilised by AI. This mirrors Agrewal et al's states of the world, but can now take any value. We then set up a second layer of judgeable characteristics, detectable only by humans. Again, this is no longer restricted to three values, but comprised of a uniform distribution that can take on a range of values. We then combine these two to form the ''adjusted health'', the true health outcome patients exhibit when they are treated.</p>

<p style="text-align: justify;">Note that we use normal and uniform distributions as examples, and that this is not based on any empirics in Agrewal et al's study suggesting that this must always be the case. Rather, we do so to show the system works with either type of distribution.</p>

```python
np.random.seed(Seed)

def base_health(mu, sigma): #Then we set up a normal distribution for the underlying health aspects, observable by AI
    dataset = np.array(sorted(pm.Normal.dist(mu, sigma).random()))
    return dataset

base_health = base_health(mu1*np.ones(n), sigma1*np.ones(n)) #we create our values for the predictable characteristics 
opportunity = np.random.uniform(low=low*np.ones(n), high=high*np.ones(n)) #and judgeable opportunities
adjusted_health = base_health + opportunity #and compile them into one
```

<p style="text-align: justify;">Next, we determine our model's basic action, which is based solely on data available for the dataset as a whole. In Agrewal et al, this is always assumed to be the safe option. Here, allow for a system that automatically determines whether it is better to treat no one (i.e., always choosing the safe option) or to treat everyone. To do so, we simply ask whether the mean of treating everyone leads to a higher health outcome than treating no one. If this is the case, we treat all patients, and if it is not, we treat no one.</p>

<p style="text-align: justify;">Additionally, we generate some data on the means and standard deviations of our health distributions.</p>

```python
Mean1 = np.average(base_health)
Mean2 = np.average(opportunity)
Mean3 = np.average(adjusted_health)

default_outcome = copy.copy(adjusted_health) #here we define our standard course of action, 
default_outcome[Mean1+Mean2<=Safe] = Safe #a baseline we can compare other outcomes too

Means1 = [Mean1, Mean2, Mean3, default_outcome.mean()] #here we compile and present some statistics on our dataset
Stds1 = [base_health.std(), opportunity.std(),adjusted_health.std(), default_outcome.std()]
Variable1 = ["Base Health", "Opportunity", "Adjusted Health", "Default Outcome"]

Table1=pd.DataFrame({"Variables":Variable1, "Means":Means1, "Standard Deviations":Stds1})
Table1=Table1[["Variables", "Means", "Standard Deviations"]]
Table1.set_index('Variables', inplace=True)
display(Table1)
```

### First Example

<p style="text-align: justify;">Next, we show how the different algorithms operate in one specific scenario, using a 50% success rate for prediction and judgement as our baselines. We first create success or failure variables for prediction and judgement, for each observation, and then measure reversions to the default option for those observations in which doing so is beneficial.
Note specifically that we could easily alter the distributions presented above, and that the algorithms automatically default to whatever the new default option is.</p>

<p style="text-align: justify;">Each algorithm utilises the binary nature of the success rate outcomes, which take up values of either 0 or 1. For each of these values, the algorithm takes a different course of action. When we have no information on predictable or judgeabable characteristics, we use the characteristic's mean value instead, taking into account its on average positive or negative contribution[$^2$](#Footnotes). This way the algorithms essentially revert to the default action in the absence of information.</p>

```python
#again, some variables we use to test

prediction_chance = 0.5
judgement_chance = 0.5

prediction_test = np.random.binomial(1,prediction_chance*np.ones(n)) #here we set up our prediction 
judgement_test = np.random.binomial(1,judgement_chance*np.ones(n)) #and judgement chances

prediction_outcome_test = copy.copy(adjusted_health) #setting up our outcomes
prediction_outcome_test[Mean2*np.ones(n) + prediction_test*base_health + (1-prediction_test)*Mean1*np.ones(n) <= Safe] = Safe    
prediction_test_clean = copy.copy(prediction_outcome_test[prediction_outcome_test != 0]) #leaning up our data

judgement_outcome_test = copy.copy(adjusted_health) #rinse and repeat
judgement_outcome_test[Mean1*np.ones(n) + judgement_test*opportunity + (1-judgement_test)*Mean2*np.ones(n)<= Safe] = Safe

judgement_test_clean = copy.copy(judgement_outcome_test[judgement_outcome_test != 0])


combined_outcome_test = copy.copy(adjusted_health)
combined_outcome_test[prediction_test*(base_health + judgement_test*opportunity\
                + (1-judgement_test)*Mean2*np.ones(n)) + (1-prediction_test)*(Mean1*np.ones(n) + judgement_test*\
                opportunity + (1-judgement_test)*Mean2) <= Safe] = Safe
combined_test_clean = copy.copy(combined_outcome_test[combined_outcome_test != 0])
```

### Generalised Cases

<p style="text-align: justify;">Next, we generalise these procedures as functions, and make them return a mean rather than individual datapoints. This way, we can compare general outcomes under a vast amount of different inputs. The various functions work similarly to the algorithms outlined above.</p>

```python
def prediction_function(p1):
    
    np.random.seed(Seed)

    prediction = np.random.binomial(1,p1*np.ones(n))

    Mean1 = np.average(base_health)
    Mean2 = np.average(opportunity)
    Mean3 = np.average(adjusted_health)

    prediction_outcome = copy.copy(adjusted_health)
    prediction_outcome[Mean2 + prediction*base_health + (1-prediction)*Mean1 <= Safe] = Safe
    Pred_Mean = [np.average(prediction_outcome)]

    return Pred_Mean

def prediction_function_std(p1):
    
    np.random.seed(Seed)

    prediction = np.random.binomial(1,p1*np.ones(n))

    Mean1 = np.average(base_health)
    Mean2 = np.average(opportunity)
    Mean3 = np.average(adjusted_health)

    prediction_outcome = copy.copy(adjusted_health)
    prediction_outcome[Mean2 + prediction*base_health + (1-prediction)*Mean1 <= Safe] = Safe
    Pred_Std = [prediction_outcome.std()]

    return Pred_Std

def judgement_function(p2):
    
    np.random.seed(Seed)

    judgement = np.random.binomial(1,p2*np.ones(n))

    Mean1 = np.average(base_health)
    Mean2 = np.average(opportunity)
    Mean3 = np.average(adjusted_health)
            
    judgement_outcome = copy.copy(adjusted_health)
    judgement_outcome[Mean1 + judgement*opportunity + (1-judgement)*Mean2<= Safe] = Safe
    Judg_Mean = [np.average(judgement_outcome)]

    return Judg_Mean

def judgement_function_std(p2):
    
    np.random.seed(Seed)

    judgement = np.random.binomial(1,p2*np.ones(n))

    Mean1 = np.average(base_health)
    Mean2 = np.average(opportunity)
    Mean3 = np.average(adjusted_health)
            
    judgement_outcome = copy.copy(adjusted_health)
    judgement_outcome[Mean1 + judgement*opportunity + (1-judgement)*Mean2<= Safe] = Safe
    Judg_Std = [judgement_outcome.std()]

    return Judg_Std



def combined_function(p1, p2):
    
    np.random.seed(Seed)

    prediction = np.random.binomial(1,p1*np.ones(n))
    judgement = np.random.binomial(1,p2*np.ones(n))

    Mean1 = np.average(base_health)
    Mean2 = np.average(opportunity)
    Mean3 = np.average(adjusted_health)

    combined_outcome = copy.copy(adjusted_health)
    combined_outcome[prediction*(base_health + judgement*opportunity\
                + (1-judgement)*Mean2) + (1-prediction)*(Mean1 + judgement*\
                opportunity + (1-judgement)*Mean2) <= Safe] = Safe
    Comb_Mean = [np.average(combined_outcome)]

    
    return Comb_Mean

def combined_function_std(p1, p2):
    
    np.random.seed(Seed)

    prediction = np.random.binomial(1,p1*np.ones(n))
    judgement = np.random.binomial(1,p2*np.ones(n))

    Mean1 = np.average(base_health)
    Mean2 = np.average(opportunity)
    Mean3 = np.average(adjusted_health)

    combined_outcome = copy.copy(adjusted_health)
    combined_outcome[prediction*(base_health + judgement*opportunity\
                + (1-judgement)*Mean2) + (1-prediction)*(Mean1 + judgement*\
                opportunity + (1-judgement)*Mean2) <= Safe] = Safe
    Comb_Std = [combined_outcome.std()]

    
    return Comb_Std


Means3 = [default_outcome.mean(), prediction_function(prediction_chance)[0], judgement_function(judgement_chance)[0], combined_function(prediction_chance, judgement_chance)[0]]
Stds3 = [default_outcome.mean(), prediction_function_std(prediction_chance)[0], judgement_function_std(judgement_chance)[0], combined_function_std(prediction_chance, judgement_chance)[0]]

Table3=pd.DataFrame({"Variables":Variable2, "Means":Means3, "Standard Deviations":Stds3})
Table3=Table3[["Variables", "Means", "Standard Deviations"]]
Table3.set_index('Variables', inplace=True)
display(Table3)
```

## Sensitivity Analysis

<p style="text-align: justify;">Now, we continue by exploring what happens when the default treatment implies the treatment of all patients, but restrict ourselves to the 3d graphs, which include the 2 2d graphs. First, we change the default outcome, and then we change the spread of judgeable characteristics. This latter case thus increases the relevance of human judgeability versus machine prediction.</p>

<a id='Graph6'></a>
### Figure 7

```python
#new variables

Seed = 1984
n = 30000
N = 100

mu2 = 2
sigma2 = 5

low2 = -5
high2 = 5

Safe = 0

np.random.seed(Seed)

#generating the new dataset

def base_health(mu, sigma):
    dataset = np.array(sorted(pm.Normal.dist(mu, sigma).random()))
    return dataset

base_health = base_health(mu2*np.ones(n), sigma2*np.ones(n))
opportunity = np.random.uniform(low=low2*np.ones(n), high=high2*np.ones(n))
adjusted_health = base_health + opportunity

def combined_function(p1, p2):
    
    np.random.seed(Seed)

    prediction = np.random.binomial(1,p1*np.ones(n))
    judgement = np.random.binomial(1,p2*np.ones(n))

    Mean1 = np.average(base_health)
    Mean2 = np.average(opportunity)
    Mean3 = np.average(adjusted_health)

    combined_outcome = copy.copy(adjusted_health)
    combined_outcome[prediction*(base_health + judgement*opportunity\
                + (1-judgement)*Mean2) + (1-prediction)*(Mean1 + judgement*\
                opportunity + (1-judgement)*Mean2) <= Safe] = Safe
    Comb_Mean = [np.average(combined_outcome)]

    
    return Comb_Mean

def combined_function_std(p1, p2):
    
    np.random.seed(Seed)

    prediction = np.random.binomial(1,p1*np.ones(n))
    judgement = np.random.binomial(1,p2*np.ones(n))

    Mean1 = np.average(base_health)
    Mean2 = np.average(opportunity)
    Mean3 = np.average(adjusted_health)

    combined_outcome = copy.copy(adjusted_health)
    combined_outcome[prediction*(base_health + judgement*opportunity\
                + (1-judgement)*Mean2) + (1-prediction)*(Mean1 + judgement*\
                opportunity + (1-judgement)*Mean2) <= Safe] = Safe
    Comb_Std = [combined_outcome.std()]

    
    return Comb_Std

#making the new graphs:

#mean

X, Y = np.meshgrid(x, y)
zs3 = np.array([combined_function(x,y) for x,y in zip(np.ravel(X), np.ravel(Y))])
Z3 = zs3.reshape(X.shape)

df3 = pd.DataFrame(Z3)
data7 = [go.Surface(z=df3.as_matrix())]

fig7 = go.Figure(data=data7, layout=layout5)


py.iplot(fig7, filename='try3')
```

## Conclusion

<p style="text-align: justify;">To illustrate the interaction between prediction by artificial intelligence and judgement by humans, we created our own data set. We chose specific test variables that were comprised of a normal and a uniform distribution that could take on different values, in contrast to what older theory assumes. </p>

<p style="text-align: justify;">With our own created dataset, we observed that average treatment outcomes increase somewhat linearly as prediction and judgement become more accurate. This applied regardless of which assumptions the model makes, and mirror earlier theory. Moreover, we demonstrated that the standard deviation generally has a marginally decreasing rate of increase, if we follow theoretical assumptions. This is the case for both prediction and judgement technologies. Additionally, we find that while increasing the prediction or judgement success rates leads to an increase in the standard deviation, their interaction behaves differently: for high success rates, further success rate increases in either the opposing variable decrease the standard deviation, and vice versa. This provides evidence of synergy between humans and machines, as demonstrated by Agrewal et al.</p>

<p style="text-align: justify;">When we relax theoretical assumptions made in earlier literature, and analyse situations in which our default decision would be to treat all patients, this changes: while mean outcomes are merely shifted up or down, the devlopment of the treatment's standard deviation undergoes a remarkable transformation, and becomes strictly decreasing. When we allow for a bigger role for human judgement, meanwhile, mean outcomes lose some of their linearity and exhibit more signs of decreasing marginal returns.
</p>

<p style="text-align: justify;">All in all, our model is capable of reproducing the results presented in earlier literature, thus proving it is sound in execution. In addition, the model allows for the relaxation of crucial assumptions, which allow for a more free-form analysis of the findings presented by previous authors. Finally, it presents these results in visual representations that are substantively more accessable than its theory-heavy predecessors, and thus helps in translating the implications found in the field of AI economics to layman terms. As such, the model above is a valuable contribution to the field of economics.
